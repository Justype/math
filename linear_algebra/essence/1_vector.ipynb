## 加减

平行四边形法则？
- 就和中学的 物理的力相加是一样的，力也是矢量嘛。

例子：

$$
\begin{bmatrix} 1\\ 2 \end{bmatrix} +
\begin{bmatrix} 3\\ -1 \end{bmatrix} =
\begin{bmatrix} 4\\ 1 \end{bmatrix}
$$

```python
v = np.array((1,2))
w = np.array((3,-1))
r = v + w           # 计算 两向量之和

ax = plt.axes()
ax.draw_vector(v,color="b") # 画出两个向量
ax.draw_vector(w,tail=v,color="r")  # 将 w的尾部移动到v
ax.draw_vector(r)
ax.ch_title("向量 数乘")
```

## 数乘

数乘，相当于对向量里面的所有数进行运算

例：

$$
-2 \times \begin{bmatrix} 1\\ 2 \end{bmatrix} =
\begin{bmatrix} -2\\ -4 \end{bmatrix}
$$

```python
v = np.array((1,2))
r = -2 * v           # 数乘

ax = plt.axes()
ax.draw_vector(v,"v", color="b") 
ax.draw_vector(r,r"$-2\times \hat v$", color="r")  
```

## 重新理解向量

个人理解：
- 一个`二维向量`是`基向量`的组合。
- `基向量`两个`线性不相关`的

### 线性相关 与 无关

`线性相关`就是 在一条直线上，或平行

`线性不相关`(Linear independent)，就是两个向量不平行：

$$
\begin{aligned}
\text{2D:}\ \  \overrightarrow w \neq a \overrightarrow v\ \ \ & a\in R \\
\text{3D:}\ \  \overrightarrow v \neq a \overrightarrow v + b \overrightarrow w\ \ \ & a,b \in R
\end{aligned}
$$

例子：

- $\begin{bmatrix} 2\\ 1 \end{bmatrix}$ 与 $\begin{bmatrix} -4\\ -2 \end{bmatrix}$  线性相关
- $\begin{bmatrix} 2\\ 1 \end{bmatrix}$ 与 $\begin{bmatrix} -2\\ 2 \end{bmatrix}$ 线性无关

```python
ax_left = plt.subplot(121) # 左边画线性相关的

v = np.array((2,1))
w = np.array((-4,-2))

ax_left.draw_vector(v, "v")
ax_left.draw_vector(w, "w")
ax_left.ch_title("线性相关")

ax_right = plt.subplot(122) # 右边画线性无关的
x = np.array((-2,2))

ax_right.draw_vector(x, "x")
ax_right.draw_vector(v, "v")
ax_right.ch_title("线性无关")
```

### 基向量 base vector

人为规定，一般是$\begin{bmatrix} 1\\ 0 \end{bmatrix}$ 和 $\begin{bmatrix} 0\\ 1 \end{bmatrix}$

其他的向量可由基向量构成

```python
i = np.array((1,0))
j = np.array((0,1))
v = -3 * i + 4 * j

ax = plt.axes()
ax.set_title(r"$\overrightarrow V = -3\hat i + 4\hat j$")
ax.draw_vector(i,"i", color='r')
ax.draw_vector(j,"j", color='b')
ax.draw_vector(v,"V")
```
